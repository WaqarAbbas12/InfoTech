Purpose: To Understand Distribution of Categorical and Continuous Variables.

## Essential Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load Dataset

```python
df = pd.read_csv(r"D:\Intership\Task-1\API_SP_POP_TOTL_DS2_en_csv_v2_9949 - Copy.csv")
df
```

## Categorical Features

### Distribution of Categorical Column Country Name from head

```python
country_names = df['Country Name'].head(30) 
year = df['2023'].head(30)  # Population data for 2023

plt.figure(figsize=(10, 6))
plt.bar(country_names, year)  # country_names on x-axis and year (population) on y-axis
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('Population in 2023')
plt.title('Population by Country in 2023')
plt.show()





```

Distribution from head is right skewed becaues majority countries have lower population resulting longer tail on right side


### Distribution from Tail

```python
country_names = df['Country Name'].tail(30) 
year = df['2023'].tail(30)  # Population data for 2023

plt.figure(figsize=(10, 6))
plt.bar(country_names, year)  # country_names on x-axis and year (population) on y-axis
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('Population in 2023')
plt.title('Population by Country in 2023')
plt.show()
```

Distrubution from tail is also right skewed majority countries have low population rate shifting tail to right side.

### Conclusion:
By Visualizing the distribution of categorical column "Country Name" for year "2023" we can conclude that the distribution of this
column for year 2023 is right skewed i.e. majority countries in this column have low population.

## Continuous Features

```python
population = df[['1960', '1961', '1962', '1963', '1964', '1965']].head(30)
sns.histplot(population)

plt.xlabel("Population")
plt.ylabel("Frequency")
plt.title('Population Distribution from 1960 to 1965')
plt.show()
```

Distribution is right skewed.

```python
population = df[['1960', '1961', '1962', '1963', '1964', '1965']].tail(30)
sns.histplot(population)

plt.xlabel("Population")
plt.ylabel("Frequency")
plt.title('Population Distribution from 1960 to 1965')
plt.show()
```

Distribution of continuous columns from 1960 to 1965 is right skewed.

## Conclusion:
We cab easily understand that he distribution of columns from 1960 to 1965 is highly right skewed meaning most
population values are concentrated near 0 show low population very few are extended to larger values.
