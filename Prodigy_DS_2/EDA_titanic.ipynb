# Data Cleaning

### Column Wise record of Missing/Null Values

```python
df.isnull().sum()
```

```python
print("Total count of Null/Missing values:",df.isnull().sum().sum())
```

# Handling Missing/Null Data

### Numeric Column

```python
df['Age'] = df['Age'].fillna(df['Age'].mode()[0])
df['Age'] = df['Age'].astype("int64") # --> Correct datatype of age. Most of the time it is integer.
```

### Categorical Column

```python
def FillNA(columns):
    for col in columns:
        df[col] = df[col].fillna(df[col].mode()[0])

columns = ['Cabin', 'Embarked']
FillNA(columns)
```

### Check Summary

It is essential to check summary to confirm that missing/null data is successfully dealt with.

```python
df.info()
print("Total count of Null/Missing values:",df.isnull().sum().sum())
```

Data is now cleaned since all null/missing values have been dealt.

# Univariate Analysis

Purpose to conduct univariate analysis to grasp the understanding of distribution and variablity of each variable indivdiually.


## PassengerID column

```python
passenger_id_counts = df['PassengerId'].value_counts()

rect = 0.1, 0.1, 0.8, 0.8
fig = plt.figure(figsize=(6, 5))

axes = fig.add_axes(rect)
axes.bar(passenger_id_counts.index, passenger_id_counts.values, color='purple')

axes.set_xlabel("Passenger_id")
axes.set_ylabel("Frequency")
axes.set_title("Distribution of PassengerID Column")
plt.tight_layout()


plt.show()
```

Distribution of PassengerID column is uniform it means PassengerID column is just an Identifier
there is no need consider this column's normality.

## Age Column

```python
age = df['Age']

plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], bins=20, kde=True, color='blue', edgecolor='black')

plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

Distribution of Age Column is right-skewed meaning most passengers are younger and very few are older stretching the tail toward right side.

## Fare Column

```python
# Boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['Fare'], color='green')
plt.title("Box Plot of Fare")
plt.xlabel("Fare")
plt.show()

```

Distribution of Fare Column is right-skewed meaning that most passengers paid relatively low fares only few high paid fares. Outliers (abnormal values) which represent passengers paid high fare to purchase luxury cabins or special tickets etc.

## Pclass Column

```python
# Bar Chart
plt.figure(figsize=(8, 6))

sns.countplot(x='Pclass', data=df, palette='Set2')
plt.title("Distribution of Pclass")
plt.xlabel("Pclass")
plt.ylabel("Frequency")
plt.legend(title="Pclass", labels=['1st Class', '2nd Class', '3rd Class'])


plt.show()
```

Th Graph represents the distribution of passenger across different classes. We can see that majority of passengers were in 3rd class followed by 

1st and 2nd class.

## Embarked Column

```python
# Bar plot for Embarked
plt.figure(figsize=(8, 6))
sns.countplot(x='Embarked', data=df, palette='Set1')

plt.title("Distribution of Embarked")
plt.xlabel("Embarked")
plt.ylabel("Frequency")
plt.legend(title='Embarked', labels=['S', 'C', 'Q'])

plt.show()
```

we can clearly see that most people embarked from port S (Southampton) followed by port C (Cherbourg) and 

port Q (Queenstown).

## Gender Column

```python
sns.countplot(data=df, x='Sex', palette="Set3")
plt.title('Gender Distribution of Passengers')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.legend(title='Gender', labels=['male', 'female'])
plt.show()
```

This Graph shows that there were more male passengers than female

# Multivariate Analysis

It is conducted to study mulitple columns and to understand their relationship with each other.

## Class, Gender, and Survival columns

```python
sns.catplot(x="Pclass", hue="Survived", col="Sex", kind="count", data=df, palette="muted")
plt.suptitle("Survival Rates by Passenger Class and Gender", y=1.05)
plt.show()
```

Plot suggests that gender played a crucial role in determining survival on the Titanic, with women having a 

higher chance of survival compared to men.

## Age, Class, and Survival

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="Age", y="Pclass", hue="Survived", data=df, palette="viridis", s=100)

plt.title("Survival by Age and Passenger Class")
plt.xlabel("Age")
plt.ylabel("Passenger Class")
plt.gca().invert_yaxis()  # keep 1st class at top

plt.show()
```

Thee plot indicates that younger passengers, especially children, had relatively higher survival rates

compared to older age groups.

## Age, Gender, and Survival

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="Sex", y="Age", hue="Survived", data=df, palette="Set2")

plt.title("Age Distribution by Gender and Survival")
plt.xlabel("Gender")
plt.ylabel("Age")

plt.show()
```

# Visualizing Fare Differences by Embarkation Port

```python
sns.boxplot(x='Embarked', y='Fare', data=df)
plt.title('Fare Distribution by Embarkation Port')
plt.xlabel('Embarkation Port')
plt.ylabel('Fare')
plt.show()
```

The graph tells that there was difference in fare based on embarkation port. Port C (Cherbourg) has high fare 

followed by port S (Southampton) and port Q (Queenstown)
